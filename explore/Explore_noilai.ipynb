```python
import re

# Example usage
text = 'Bây giờ mày ra đây đi nói cho rõ ra là mày muốn cái gì nghiền ngẫm cho kĩ thiệt kĩ vào'
substrings = sorted(vietnamese_grammar['phụ âm'] + vietnamese_grammar['phụ âm ghép'], key=lambda x:(-len(x),x))

def clean_phu_am(text):
    pattern = '|'.join(map(re.escape, substrings))
    filtered_text = re.sub(pattern, '', text.lower())
    return filtered_text

print()
```

```python
set_van_co_dau = {}
for word,count in n_gram_phrases.items():
    van_co_dau = clean_phu_am(word).split()
    for van in van_co_dau:
        if (clean_mark(van), van) not in set_van_co_dau:
            set_van_co_dau[(clean_mark(van), van)] = 0
        set_van_co_dau[(clean_mark(van), van)] += count
```
